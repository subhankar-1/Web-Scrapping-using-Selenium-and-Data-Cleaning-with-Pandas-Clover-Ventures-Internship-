```python
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from datetime import date
from selenium.webdriver.common.keys import Keys
import os
import time
import pandas as pd

sku = 'sku_blah'
quantity = 'quantity_blah'
price = 'price_blah'
location = 'Mumbai'
source = 'Natures_Basket'

csvList = []
csvDict = {
        'sku': sku,
        'quantity': quantity,
        'price': price,
        'date': date,
        'location': location,
        'source': source,
}

today = date.today()
d1 = today.strftime("%Y-%m-%d")
date_1 = today.strftime("%d_%m_%Y")

if os.path.exists('nature.csv') == True:
    print('This file exists, not writing to the database')
else:
    print('This file does not exist and hence, writing to the database')

    url = "https://www.naturesbasket.co.in/Online-grocery-shopping/Fruits-Vegetables/Daily-Vegetables/34_0_0"

    browser = webdriver.Chrome("/home/subhankar/Downloads/chromedriver")
    browser.get(url)
    browser.maximize_window()
    pincode_selector = browser.find_element_by_id("ctl00_txtSearchServicablePin")
    pincode_selector.click()
    time.sleep(5)
    pincode_text = browser.find_element_by_id("txt")
    pincode_text.send_keys("400102")
    time.sleep(2)
    pincode_text.send_keys(Keys.ENTER)
    time.sleep(5)
    browser.get(url)
    scroll_down = browser.find_element_by_tag_name("html")
    scroll_down.send_keys(Keys.END)
    time.sleep(4)
    scroll_down.send_keys(Keys.HOME)
    time.sleep(4)
    res = browser.execute_script("return document.documentElement.outerHTML")
    page_soup = soup(res, "html.parser")
    containers = page_soup.findAll("div", {"productbox": "1"})

    for container in containers:

        header = container.findAll("a", {"class": "search_Ptitle linkdisabled"})
        head = header[0].text

        try:
            try:
                quantity_container = container.findAll("div", {"class":"search_PSelectedSize"})
                quantity = quantity_container[0].text
                #quant = q.split()
                #quantity = quant[0]

            except:
                quantity_container = container.findAll("span", {"class": "search_PSelectedSize"})
                quantity = quantity_container[0].text
                #quant = q.split()
                #quantity = quant[0]

        except:
                quantity_container = container.findAll("div", {"class": "outofstock"})
                quant = quantity_container[0].text
                if quant == "Sold out":
                    quantity = '0'

                else:
                    quantity = quant

        price_container = container.findAll("div", {"class":"productlist-price"})
        price_1 = price_container[0].text.strip()
        price = price_1.split("â‚¹")

        csvDict['sku'] = str(head).replace(',','')
        csvDict['price'] = str(price[1]).replace(',','')
        csvDict['quantity'] =str(quantity).replace(',','')
        csvDict['date'] = d1
        csvDict['location'] = location
        csvDict['source'] = source

        csvList.append(csvDict.copy())

    print("the no of items is ",len(csvList))
    print(csvList)

    browser.close()
    nature = open("nature_MUM_" + date_1 + ".csv", "w")



    def write_to_csv(output):
        with open("nature_MUM_" + date_1 + ".csv", "w", encoding="utf-8") as f:
            f.write("sku,price,Quantity,date,location,source\n")
            for line in output:
                f.write(f"{line['sku']},{line['price']},{line['quantity']},{line['source']},{line['date']},{line['location']}\n")


    nature.close()
    write_to_csv(csvList)
    df=pd.read_csv("nature_MUM_" + date_1 + ".csv")
    df.to_csv(date_1+'.csv', mode='a', header=False,index=False)
    
```
