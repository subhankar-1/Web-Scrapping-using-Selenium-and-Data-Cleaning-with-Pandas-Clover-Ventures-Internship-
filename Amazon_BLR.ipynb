```python
from bs4 import BeautifulSoup as soup
from datetime import date
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
import time
import pandas as pd

sku = 'sku_blah'
quantity = 'quantity_blah'
price = 'price_blah'
location = "Bangalore"
source = 'Amazon'

csvList = []
csvDict = {
    'sku': sku,
    'quantity': quantity,
    'price': price,
    'date': date,
    'location': location,
    'source': source,
}

today = date.today()
d1 = today.strftime("%Y-%m-%d")
date_1 = today.strftime("%d_%m_%Y")

if os.path.exists(r'amazon.csv') == True:
    print('This file exists, not writing to the database')
else:
    print('This file does not exist and hence, writing to the database')

    #browser = webdriver.Firefox()
    browser = webdriver.Chrome("/home/subhankar/Downloads/chromedriver")
    browser.maximize_window()
    amazon_landing_page = "https://www.amazon.in/"
    browser.get(amazon_landing_page)
    landing_page = browser.find_element_by_tag_name("html")
    selector = browser.find_element_by_id("nav-packard-glow-loc-icon")
    selector.click()
    input_text = browser.find_element_by_class_name("a-popover-inner").click()
    time.sleep(2)
    pincode = browser.find_element_by_id('GLUXZipUpdateInput').send_keys("560002", Keys.ENTER)
    time.sleep(3)
    my_url = ["https://www.amazon.in/s?i=grocery&bbn=16392737031&rh=n%3A4859731031&brr=1&pf_rd_p=4f9a08b4-56c1-4896-9456-bd06bce4293a&pf_rd_r=4VVJFQKTMJEQY9PZK178&qid=1571375599&rd=1&ref=sr_pg_1",
              "https://www.amazon.in/s?i=grocery&bbn=16392737031&rh=n%3A4859731031&page=2&brr=1&pf_rd_p=4f9a08b4-56c1-4896-9456-bd06bce4293a&pf_rd_r=4VVJFQKTMJEQY9PZK178&qid=1571467795&rd=1&ref=sr_pg_2",
              "https://www.amazon.in/s?i=grocery&bbn=16392737031&rh=n%3A4859731031&page=3&brr=1&pf_rd_p=4f9a08b4-56c1-4896-9456-bd06bce4293a&pf_rd_r=4VVJFQKTMJEQY9PZK178&qid=1571467795&rd=1&ref=sr_pg_3",
              "https://www.amazon.in/s?i=grocery&bbn=16392737031&rh=n%3A4859731031&page=4&brr=1&pf_rd_p=4f9a08b4-56c1-4896-9456-bd06bce4293a&pf_rd_r=4VVJFQKTMJEQY9PZK178&qid=1571467795&rd=1&ref=sr_pg_4",
              "https://www.amazon.in/s?i=grocery&bbn=16392737031&rh=n%3A4859731031&page=5&brr=1&pf_rd_p=4f9a08b4-56c1-4896-9456-bd06bce4293a&pf_rd_r=4VVJFQKTMJEQY9PZK178&qid=1571467795&rd=1&ref=sr_pg_5"

              ]
    for url in my_url:
        browser.get(url)
        #add the ',' replacement everywhere before extacting the content
        res = browser.execute_script("return document.documentElement.outerHTML")
        page_soup = soup(res, "html.parser")
        containers = page_soup.findAll("div", {"class": "a-section a-spacing-medium"})
        for container in containers:
            try:
                try:
                    header_container = container.findAll( "span",
                                                          {"class": "a-size-base-plus a-color-base a-text-normal"} )
                    header = header_container[0].text
                    quantity_contain = header.split( "," )
                    head = quantity_contain[0]
                    q = quantity_contain[2]
                    quant = q.split()
                    quantity = quant[0]

                except:
                    q = quantity_contain[1]
                    quant = q.split()
                    quantity = quant[0]

            except:
                try:
                    quantity_contain = header.split( "-" )
                    q = quantity_contain[1]
                    quant = q.split()
                    quantity = quant[0]

                except:
                    quantity_contain = header.split( "-" )
                    quantity = "check"

            try:
                price_header = container.findAll( "span", {"class": "a-price-whole"} )
                price = price_header[0].text
            except:
                price = '0'

            csvDict['sku'] = head.replace(',','')
            csvDict['price'] = price.replace(',','')
            csvDict['quantity'] = quantity.replace(',','')
            csvDict['date'] = d1
            csvDict['location'] = location

            csvList.append(csvDict.copy())

    print("the no of items is", len(csvList))
    print(csvList)
    browser.close()
    amazon_1 = open("amazon_BNG_" + date_1 + ".csv", "w")


    def write_to_csv(output):
        with open("amazon_BNG_" + date_1 + ".csv", "w",  encoding="utf-8") as f:
            f.write("sku,price,Quantity,date,location,source\n")
            for line in output:
                f.write(f"{line['sku']},{line['price']},{line['quantity']},{line['source']},{line['date']},{line['location']}\n")


    amazon_1.close()
    write_to_csv(csvList)
    df=pd.read_csv("amazon_BNG_" + date_1 + ".csv")
    df.to_csv(date_1+'.csv', mode='a', header=False,index=False)
    
```
