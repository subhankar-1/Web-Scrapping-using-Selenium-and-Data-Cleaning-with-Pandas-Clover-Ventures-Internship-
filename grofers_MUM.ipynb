```python
import csv
import time
import pandas as pd
from datetime import date
from selenium import webdriver
from bs4 import BeautifulSoup as soup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from lxml import html
import re 
import numpy as np
def use_selenium_automation():
    driver =webdriver.Chrome("/home/subhankar/Downloads/chromedriver")
    driver.maximize_window()
    return driver
def crawl_and_scrap_webpages_yt(driver,url):
    driver.get(url) 
    extracted_page_source = driver.page_source
    time.sleep(1)
    try:
        city = driver.find_element_by_xpath("/html/body/div[1]/div/div[2]/div[2]/header/div[2]/div[2]/div/div/div/div[2]/div/div[3]/div[1]/div/div/div/div/div[2]/input") 
        city.send_keys("Mumbai") 
        city.send_keys(Keys.ENTER)
        time.sleep(3)
        f_v=driver.find_element_by_partial_link_text("Vegetables & Fruits")
        f_v.send_keys(Keys.ENTER)
        time.sleep(3)
        last_height = driver.execute_script("return document.documentElement.scrollHeight")
        driver.execute_script("window.scrollTo(0,document.documentElement.scrollHeight);")
        time.sleep(2)
        while True:
            driver.execute_script("window.scrollTo(0,document.documentElement.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.documentElement.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        time.sleep(2)
        name=driver.find_elements_by_xpath('//div[@class="plp-product__name "]')
        names=[x.get_attribute("title").replace(',','') for x in name]
        time.sleep(2)
        quantity=driver.find_elements_by_xpath('//div[@class="plp-product__quantity"]')
        quantities=[x.get_attribute("title").replace(',','') for x in quantity]
        time.sleep(2)
        price=driver.find_elements_by_xpath('//span[@class="plp-product__price--new"]')
        prices=[x.text.replace(',','') for x in price]
        return names, quantities, prices
    except:
        print("error")
        return [],[],[]
if __name__=="__main__":
    today = date.today()
    d1 = today.strftime("%Y-%m-%d")
    date_1 = today.strftime("%d_%m_%Y")
    driver=use_selenium_automation()  
    urls=["https://grofers.com/categories"]
    for url in urls:
        name, quantity, price = crawl_and_scrap_webpages_yt(driver,url)
    driver.close()
    name_dict = {'Name': name,
            'Price(Rs)': price,
            'Quantity': quantity
          }
    df = pd.DataFrame(name_dict)
    df['Website']='grofers'
    df['date']=d1
    df['location']='mumbai'
    df
    #df.to_csv('grfers_DELHI.csv',index=False)
    df.to_csv(date_1+'.csv', mode='a', header=False,index=False)
```
