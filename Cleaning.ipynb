```python
df=df.rename(columns={"price": "Price(Rs)","sku":"Name","quantity":"Quantity"})
df['Name'].update(df.Name.str.split(r"([0-9])",n=1,expand=True)[0])
df.loc[df['Name'].str.contains('kg', case=False), 'Quantity'] = '1 kg'
df['Name'] = df.Name.apply(lambda x: x.replace('Spar',''))#'-','SPAR','Fresh'
df['Name'] = df.Name.apply(lambda x: x.replace('SPAR',''))
df['Name'] = df.Name.apply(lambda x: x.replace('Fresh',''))
df['Name'] = df.Name.apply(lambda x: x.replace('-',''))
to_rep=['Kg','Pc','Box','Grofers',"Namdhari","Delicious","delicious",'Fruit & Vegetable']
for st in to_rep:
    df['Name'] = df.Name.apply(lambda x: x.replace(st,''))
df
```

```python
df['Quantity'] = df.Quantity.astype(str).apply(lambda x: x.replace(x[0:x.find('%')+1],''))
df['Quantity'] = df.Quantity.astype(str).apply(lambda x: x.replace(x[x.find('('):x.find(')')+1],''))
df1 =  df['Quantity'].str.extractall('(?P<a>\d+\.\d+|\d+)\s*(?P<b>[A-Za-z%]*)')
```

```python
df1
```

```python
d = {'gm':0.001,'gr':0.001,'grams':0.001,'kg':1,'g':0.001,'unit':1,'units':1,'ml':1,'GM':0.001,'G':0.001,'Ml':1,'Gm':0.001,'gp':0.001,'Pb':0.001,'Pc':1,'Gpb':0.001,'pcs':1,'pc':1,'NaN':1,'':1,'nan':1,'GMS':0.001,'N':1,'PIECE': 1,'KG' :1,'Kg':1,'PIECES':1,'Bunch':1,'Items':1,'items':1,'Item':1,'item':1,'Bunches':1,'bunch':1,'Bunch':1}
a = df1['a'].astype(float).mul(df1['b'].astype(str).map(d)).unstack().prod(axis=1)                                                              #check for Pb
print (a)
#df1['b'].astype(str).map(d)
```

```python
b= df1['b'].astype(str)
df1['b']=df1['b'].replace(to_replace = ['g', 'gm','G','Gm','Gpb','Gp','Pb','GM','GMS','KG','Kg'], value = 'kg')
df1['b']=df1['b'].replace(to_replace = ['N', 'PIECE','PIECES','unit','Pc','pc','pcs','Items'], value = 'units')
df=df.rename(columns={"Price(Rs)": "Price"})

df['Price']=df.Price.str.extract('(\d*\.\d+|\d+)')
df = df[df['Price'].astype(str) != '0']
df['Price'] = pd.to_numeric(df['Price'])
df['Price']=(df['Price']/a).round(2)
b= "1 " + b
df['Quantity']="1 "+df1['b'].unstack()
df['Quantity'] = df['Quantity'].fillna("1 units")
```
