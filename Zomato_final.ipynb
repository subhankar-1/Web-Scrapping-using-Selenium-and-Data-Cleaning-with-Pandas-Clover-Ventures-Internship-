```python
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from datetime import date
import os
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from lxml import html

zomato_list = []
today = date.today()
d1 = today.strftime( "%Y-%m-%d" )
date_1 = today.strftime( "%d_%m_%Y" )
browser = webdriver.Firefox()
browser.maximize_window()
zomato_landing_page = "https://www.zomato.com/bangalore/restaurants?page=1"
browser.get( zomato_landing_page )
res = browser.execute_script( "return document.documentElement.outerHTML" )
page_soup = soup( res, "lxml" )
page_range = page_soup.findAll( "div", {"class": "col-l-4 mtop pagination-number"} )[0].text
page_no = page_range.split( "of" )[-1]
total_page = int( page_no )
#print(total_page)
with open( "zomato_restaurants" + date_1 + ".csv", "w" ) as f:
    f.write( "restaurant_name,locality,address,cuisines\n" )
for page in range(1,total_page):
    response = browser.get( "https://www.zomato.com/bangalore/restaurants?page={}".format( page ) )
    search_list =browser.find_elements_by_xpath("//*[@class='card  search-snippet-card     search-card  ']")
    head_contain=browser.find_elements_by_xpath( "//*[@data-result-type='ResCard_Name']")
    head_contain = [x.text.replace(',','') for x in head_contain]
    head_contain = [x.replace('\n','') for x in head_contain]
    locality = browser.find_elements_by_tag_name("b")
    locality = [x.text.replace(',','') for x in locality]
    locality = [x.replace('\n','') for x in locality]
    address = browser.find_elements_by_xpath( "//*[@class='col-m-16 search-result-address grey-text nowrap ln22']")
    address = [x.text.replace(',','') for x in address]
    address = [x.replace('\n','') for x in address]
    rest2 = browser.find_elements_by_xpath( "//*[@class='col-s-11 col-m-12 nowrap  pl0']")
    rest2 = [x.text.replace('\n','') for x in rest2]
    zomato_dict = {"restaurant_name":head_contain ,
            "locality": locality[1:-2],
            'address': address ,
            'cuisines': rest2
    }
    #print(len(head_contain))
    #print(len(locality[1:-2]))
    #print(len(address))
    #print(len(rest2))
    df= pd.DataFrame( zomato_dict )
    df.to_csv( "zomato_restaurants" + date_1 + ".csv", mode='a', header=False,index=False)
df =pd.read_csv( "zomato_restaurants" + date_1 + ".csv")
new = df['cuisines'].str.split(",", expand = True) 
df=df.drop(columns=['cuisines'],axis=1)
for col in new.columns:
    df['cuisines_'+str(col)]=new[col]
df=df.astype(str).replace('None',' ')
df=df.drop_duplicates()
df.to_csv( "zomato_restaurants" + date_1 + ".csv",index=False)
browser.close()
```
