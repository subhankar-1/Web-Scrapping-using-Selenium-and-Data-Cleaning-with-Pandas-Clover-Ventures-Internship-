```python
import csv
import datetime
import os
import pandas as pd
import xlsxwriter
import numpy as np
from datetime import date
now = datetime.datetime.now()
today = datetime.date.today()
strtoday = now.strftime('%d-%m-%Y')
yesterday = today-datetime.timedelta(days=1)
yday = yesterday.strftime('%d-%m-%Y')
d1=date.today()
```

```python
df_cus = pd.read_excel("customer_list.xlsx" )
df_os= pd.read_excel("order_summary.xlsx" )
df_io= pd.read_excel("items-wise-order-transactions.xlsx" )
```

```python
df_os['Created']=pd.to_datetime(df_os['Created'],format='%Y-%m-%d %H:%M:%S' ).dt.strftime('%d-%m-%Y')
df_os=df_os[df_os['Created']==(d1-datetime.timedelta(days=1)).strftime("%d-%m-%Y")]
df_os=df_os[df_os['Order State']!='Cancelled']
```

```python
df_io['Order date-time']=pd.to_datetime(df_io['Order date-time'],format='%Y-%m-%d %H:%M:%S' ).dt.strftime('%d-%m-%Y')
df_io=df_io[df_io['Order date-time']==(d1-datetime.timedelta(days=1)).strftime("%d-%m-%Y")]
df_io=df_io[df_io['Order State']!='Cancelled']
df_date=df_io['Order date-time']
```

```python
df_cus=df_cus[['User ID','Name','Mobile']]
df_os=df_os[['User Id','Id','Address','Pin']]
df_io=df_io[['Order ID','Item Name','Quantity','Total Price']]
df_cus=df_cus.rename(columns={'User ID':'User Id'})
```

```python
df_d2c=pd.merge(df_os,df_cus,on='User Id',how='left')
```

```python
df_d2c=df_d2c.drop('User Id',axis=1)
df_d2c=df_d2c.rename(columns={'Id':'Order ID'})
```

```python
df_d2c=pd.merge(df_io,df_d2c,on='Order ID',how='left')
```

```python
df_data = pd.DataFrame(columns = ["Date","customer_name","address","Pin","mobile_number","sku","qty","amount"])
df_data['customer_name']=df_d2c['Name']
df_data['address']=df_d2c['Address']
df_data['mobile_number']=df_d2c['Mobile']
df_data['sku']=df_d2c['Item Name']
df_data['Pin']=df_d2c['Pin']
df_data['qty']=df_d2c['Quantity']
df_data['amount']=df_d2c['Total Price']
df_data['Date']=(d1-datetime.timedelta(days=2)).strftime("%d-%m-%Y")
try:
    new = df_data["address"].str.split("-", n = 1, expand = True) 
    df_data['address']=new[1]
except:
    ;
```

```python
total_customer=len(df_data['mobile_number'].unique())
total_order_amount=df_data['amount'].sum()
df_cus = pd.read_excel("customer_list.xlsx" )
df_cp=pd.DataFrame(columns = ["Pin","No. of customers"])
df_data['Pin']=df_data['Pin'].fillna("not defined")
df_cp['Pin']=pd.Series(df_data['Pin'].unique())
for pin in df_cp['Pin'].unique():
    df_cp.loc[df_cp['Pin']==pin,'No. of customers']=len(df_data[df_data['Pin']== pin]['mobile_number'].unique())
#df_cp=df_cp.dropna()
```

```python
df_new_cus = pd.DataFrame(columns = ["User ID","customer_name","address","mobile_number","email"])
df_new_cus['mobile_number']=pd.Series(df_new['mobile_number'].unique())
for no in df_new_cus['mobile_number'].unique():
    df_new_cus.loc[df_new_cus['mobile_number']==no,'User ID']=df_cus[df_cus['Mobile']== no].iloc[0,0]
    df_new_cus.loc[df_new_cus['mobile_number']==no,'customer_name']=df_cus[df_cus['Mobile']== no].iloc[0,3]
    df_new_cus.loc[df_new_cus['mobile_number']==no,'address']=df_data[df_data['mobile_number']== no].iloc[0,2]
    df_new_cus.loc[df_new_cus['mobile_number']==no,'email']=df_cus[df_cus['Mobile']== no].iloc[0,5]
```
