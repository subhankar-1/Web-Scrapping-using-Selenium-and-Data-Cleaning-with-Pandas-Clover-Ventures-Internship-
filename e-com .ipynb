```python
import csv
import datetime
import os
import pandas as pd
import xlsxwriter
import numpy as np
from datetime import date
```

```python
now = datetime.datetime.now()
today = datetime.date.today()
strtoday = now.strftime('%d-%m-%Y')
yesterday = today-datetime.timedelta(days=1)
yday = yesterday.strftime('%d-%m-%Y')
```

```python
df = pd.read_csv(yday+" SKU Data.csv" )
```

```python
df
```

```python
df['key']=df['Order ID'].astype(str)+df['Order date-time'].astype(str)
```

```python
df_del=pd.read_csv(yday+" Delivery Data.csv" )
df_del=df_del[['ID','Created At','Request Delivery Time','Time Slot Start','Time Slot End']]
df_del['key']=df_del['ID'].astype(str)+df_del['Created At'].astype(str)
df_del=df_del.drop(columns=['ID','Created At'],axis=1)
df_del
```

```python
df=pd.merge(df,df_del,on='key',how='left')
```

```python
df_add=pd.read_csv(yday+" Address Data.csv" )
df_add=df_add[['Id','Created','Address','Sub Locality','Pin']]
df_add['key']=df_add['Id'].astype(str)+df_add['Created'].astype(str)
df_add=df_add.drop(columns=['Id','Created'],axis=1)
df_add
```

```python
df=pd.merge(df,df_add,on='key',how='left')
```

```python
df_data = pd.DataFrame(columns = ["Order ID","Customer Name","Order date-time","Item Name","Quantity","Unit Price","Total Price","Vehicle"])
```

```python
df_data['Order ID']=df['Order ID']
df_data['Customer Name']=df['Customer Name']
df_data['Order date-time']=df['Order date-time']
df_data['Item Name']=df['Item Name']
df_data['Quantity']=df['Quantity']
df_data['Unit Price']=df['Unit Price']
df_data['Total Price']=df['Total Price']
df_data['Vehicle']=' '
```

```python
df_conv=pd.read_excel("Conversion_file.xlsx",'Sheet1' )
```

```python
df_conv=df_conv[['Ecommerce SKU','Tally sku','Ecommerce Conversion','Clover conversion']]
df_conv=df_conv.rename(columns={'Ecommerce SKU':'Item Name'})
df_conv
```

```python
df_data=pd.merge(df_data,df_conv,on='Item Name',how='left')
df_data
```

```python
df_data['QTY Ecom']=df_data['Quantity']*df_data['Ecommerce Conversion']
df_data=df_data.drop('Clover conversion',axis=1)
```

```python
df_route= pd.DataFrame(columns = ["Order ID","Customer Name",'Address','Sub Locality','Request Delivery Time','Time Slot Start','Time Slot End',"Vehicle",'Location'])
```

```python
df_route['Order ID']=df['Order ID']
df_route['Customer Name']=df['Customer Name']
df_route['Address']=df['Address']
df_route['Sub Locality']=df['Sub Locality']
df_route['Request Delivery Time']=df['Request Delivery Time']
df_route['Time Slot Start']=df['Time Slot Start']
df_route['Time Slot End']=df['Time Slot End']
df_route['Vehicle']=' '
apar_list=pd.read_excel('Apartment list.xlsx','Apartment list')
apar_list=apar_list.iloc[:,0]
apar_list=apar_list.dropna()
apar_list=apar_list.tolist()
def map_substring(s, dict_map):
    for key in dict_map:
        if key.lower().replace(' ','') in str(s).lower().replace(' ',''): return key
    return np.nan
df_route['Location'] = df_route['Address'].apply(lambda x: map_substring(x, apar_list))
```

```python
df_route=df_route.drop_duplicates()
```

```python
df_route
```

```python
df=df.drop('key',axis=1)
writer = pd.ExcelWriter(yday+" SKU Data Workbook.xlsx",
        engine='xlsxwriter' )
workbook = writer.book
df.to_excel( writer, sheet_name='Master Data',index=False )
df_data.to_excel( writer, sheet_name='Data' ,index=False)
df_route.to_excel( writer, sheet_name='Route' ,index=False)
writer.save()
```
